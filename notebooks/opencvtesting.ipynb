```python
import cv2 as cv
import numpy as np
```

```python
img_var = cv.imread('../data/images/blueRose.jpeg')
```

```python
if img_var is not None:
    cv.imshow('windowss', img_var)
    cv.waitKey(0)            # <- required to process events and keep window open
    cv.destroyAllWindows()
```

```python
capture = cv.VideoCapture(0)
while True:
    isTrue, frame_var = capture.read()
    if isTrue:
        cv.imshow('Webcam Video', frame_var)
    else:
        break
    if cv.waitKey(20) & 0xFF==ord('q'):
        break
```

```python
img_var = cv.resize(img_var, [250,250], cv.INTER_CUBIC)
```

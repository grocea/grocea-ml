```python
import cv2 as cv
import pytesseract
from matplotlib import pyplot as plt

#if TesseractNotFoundError occurs
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\ammar.jamalludin\AppData\Local\Programs\Tesseract-OCR\tesseract.exe'
```

get image

```python
image_path = "../data/images/tesseract-image.png"

image = cv.imread(image_path)

image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

convert to grayscale

```python
gray = cv.cvtColor(image_rgb, cv.COLOR_RGB2GRAY)
print("Grayscale Image")
# cv.imshow( 'window',gray)
# cv.waitKey(0)            # <- required to process events and keep window open
# cv.destroyAllWindows()
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
extracted_text = pytesseract.image_to_string(image_rgb)
print(" Extracted Text:\n")
print(extracted_text)
```

```python
data = pytesseract.image_to_data(image_rgb, output_type=pytesseract.Output.DICT)

n_boxes = len(data['level'])
for i in range(n_boxes):
    (x, y, w, h) = (data['left'][i], data['top'][i], data['width'][i], data['height'][i])
    cv.rectangle(image_rgb, (x, y), (x + w, y + h), (255, 0, 0), 2)
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(image_rgb)
plt.title("Image with Text Bounding Boxes")
plt.axis("off")
plt.show()
```
